```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob
import cv2
```

```python
#http://blog.ayoungprogrammer.com/2013/04/tutorial-detecting-multiple-rectangles.html/
img_dir = [y for x in os.walk('C:/Git/NetData/Set3') for y in glob(os.path.join(x[0], '*.png'))]
```

```python
# load the image and resize it to a smaller factor so that
# the shapes can be approximated better
img_idx = 200
img = cv2.imread(img_dir[img_idx], 0)
```

```python
img.shape
```

```python
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
lsd = cv2.createLineSegmentDetector(0)
```

```python
lines = lsd.detect(img)[0]
```

```python
if lines is not None:
    drawn_img = lsd.drawSegments(img,lines)
else:
    drawn_img = img
```

```python
plt.figure(figsize=(20,10))
plt.axis("off")
plt.imshow(cv2.cvtColor(drawn_img, cv2.COLOR_BGR2RGB))
plt.show()
```
